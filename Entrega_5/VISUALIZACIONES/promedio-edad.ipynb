```python
from google.colab import files
import pandas as pd
import plotly.express as px

uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded)))
```

```python
df.columns = [c.lower() for c in df.columns]

avg_age_per_continent = df.groupby('continente')['pro. de edad'].mean().reset_index()
avg_age_per_continent.rename(columns={'pro. de edad': 'avg_continente_edad'}, inplace=True)

df_merged = pd.merge(df, avg_age_per_continent, on='continente', how='left')

fig = px.sunburst(
    df_merged,
    path=["continente", "país"],
    values="pro. de edad",
    color="pro. de edad",
    custom_data=["continente", "país", "pro. de edad", "avg_continente_edad"],
    color_continuous_scale="Viridis"
)

fig.update_layout(
    title="Promedio de edad por continente y país"
)

new_hovertemplates = []
ids = fig.data[0].ids
continents_list = df['continente'].unique().tolist()

for i, node_id in enumerate(ids):
    if node_id == 'total' or node_id == '':
        hovertemplate = "<b>Total</b><br>Hover over a continent or country for details<extra></extra>"
    elif node_id in continents_list:
        hovertemplate = (
            "<b>Continente</b>: %{label}<br>" +
            "<b>Promedio de edad</b>: %{customdata[3]:.3f}<extra></extra>"
        )
    else:
        hovertemplate = (
            "<b>País</b>: %{label}<br>" +
            "<b>Promedio de edad</b>: %{customdata[2]:.3f}<br>" +
            "<b>Continente</b>: %{customdata[0]}<extra></extra>"
        )
    new_hovertemplates.append(hovertemplate)

fig.data[0].hovertemplate = new_hovertemplates

fig.show()
```

```python
fig.write_html("sunburst_chart.html")
print("Interactive sunburst chart saved as 'sunburst_chart.html'")
```
